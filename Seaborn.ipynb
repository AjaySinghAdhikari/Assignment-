```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri= sns.load_dataset("fmri")
```

```python
sns.lineplot(data= fmri, x="timepoint", y ="signal", hue="event", style = "region")
plt.show
```

```python
sns.lineplot(data= fmri, x="timepoint", y ="signal", hue="event", style = "region")
sns.barplot(data= fmri, x="timepoint", y ="signal", hue="event")
sns.boxplot(data= fmri, x="timepoint", y ="signal", hue="event")
sns.histplot(data= fmri, x="timepoint", y ="signal", hue="event")
sns.heatmap(data= fmri, x="timepoint", y ="signal", hue="event")
```

```python
titanic = sns.load_dataset('titanic')
```

```python
diamonds= sns.load_dataset("diamonds")
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.pairplot(data = iris, hue ="species")
```

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
```

```python
plt.figure(figsize =(10,6))
sns.heatmap(data= flights_pivot, cmap="coolwarm",annot =True, fmt="d")
plt.title("Flight Passenger Heatmap")
plt.show()
```
