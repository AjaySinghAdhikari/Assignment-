```python
"""Matplotlib:
Matplotlib is a popular Python library for data visualization.
It is mainly used for creating static, animated, and interactive visualizations.
Why is it used?
It provides a variety of visualization tools for exploring and presenting data.
It helps understand trends, patterns, and relationships in data.
Five Plots in Pyplot:
Line Plot (plt.plot()) – Shows trends over time.
Scatter Plot (plt.scatter()) – Displays the relationship between two numerical variables.
Bar Plot (plt.bar()) – Compares categories using bars.
Histogram (plt.hist()) – Shows the distribution of numerical data.
Box Plot (plt.boxplot()) – Displays data distribution and outliers."""
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
np.random.seed(3)
x = 3+ np.random.normal(0,2,50)
y = 3+ np.random.normal(0,2,len(x))
```

```python
plt.figure(figsize =(1.8,1.5))
plt.scatter(x,y, color = "blue", alpha = 0.9)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(1,1,1)

plt.plot(x,y, color ="blue")
```

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(x,y2, marker ="o", color ="r")

plt.subplot(2,2,2)
plt.plot(x,y3, marker ="d", color ="b")


plt.subplot(2,2,3)
plt.plot(x,y4, marker ="x", color ="g")


plt.tight_layout()
plt.show()





```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10,6))
plt.subplot(2,2,2)
plt.bar(company,profit, color = ['red','blue','green','purple'])
plt.show()

plt.subplot(1,2,1)
plt.barh(company,profit, color = ['red','blue','green','purple'])
plt.show()


```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(10,5))
plt.subplot(2,2,2)
plt.boxplot([box1,box2],labels =["Box 1","Box 2"], patch_artist= True)
plt.show()
```
